# 1. Metodo Monte carlo

Las simulaciones de Monte Carlo son una técnica matemática que predice los posibles resultados de un evento incierto. Los programas informáticos utilizan este método para analizar datos pasados y predecir una serie de resultados futuros en función de una elección de acción.

# Introducción
El método Monte Carlo (MMC) es una técnica numérica basada en el uso de números aleatorios para obtener soluciones aproximadas a problemas matemáticos complejos y sistemas físicos. Aunque el nombre "Monte Carlo" evoca imágenes del famoso casino en Mónaco, en realidad, el método debe su nombre a las simulaciones que se realizan, las cuales son análogas a los juegos de azar.

El método se utiliza en una amplia variedad de campos, desde la física estadística hasta la optimización financiera y la inteligencia artificial. Su esencia radica en utilizar muestreos aleatorios o técnicas de muestreo estocástico para obtener resultados numéricos para problemas que podrían ser deterministas en principio.

# Principio Básico
El principio detrás del MMC es simple: en lugar de tratar de analizar un sistema utilizando ecuaciones complicadas y soluciones analíticas, se simula el sistema un gran número de veces utilizando números aleatorios y luego se toma un promedio (o algún otro estadístico) de las simulaciones para obtener una respuesta aproximada.

Por ejemplo, para estimar el valor de $\pi$, se podría simular lanzamientos aleatorios de dardos en un cuadrado que inscribe un círculo. La proporción de dardos que caen dentro del círculo en comparación con el total de dardos lanzados puede usarse para estimar el área del círculo y, por ende, el valor de $\pi$.


La aproximación de π usando el método de Monte Carlo se basa en la relación entre el área de un círculo de radio r = 1 y el área de un cuadrado de lado l = 2. Si lanzamos aleatoriamente puntos dentro del cuadrado y contamos cuántos caen dentro del círculo, podemos estimar π.

La relación de áreas es:
\begin{equation}
\frac{\text{Área del círculo}}{\text{Área del cuadrado}} = \frac{\pi r^2}{l^2}
\end{equation}

Para (r = 1\) y (l = 2), esta relación es ${\pi}/{4}$ Por lo tanto, el valor de π se puede estimar como:

\begin{equation}
\pi \approx 4 \times \frac{\text{Número de puntos dentro del círculo}}{\text{Número total de puntos lanzados}}
\end{equation}







```python
# Importamos las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt

# Establece cuántos puntos vamos a crear
N = 1 #!

# ---- CREACIÓN DE PUNTOS ALEATORIOS ----
# Genera coordenadas x aleatorias entre -1 y 1
x = np.random.uniform(-1, 1, N)
# Genera coordenadas y aleatorias entre -1 y 1
y = np.random.uniform(-1, 1, N)

# Lista para guardar las estimaciones de pi en cada iteración
pi_estimations = []

# Inicializamos un contador para saber cuántos puntos están dentro del círculo
inside_count = 0

# ---- PROCESAMIENTO DE PUNTOS ----
# Recorremos cada punto
for i in range(N):
    # Si el punto está dentro del círculo (distancia al origen <= 1), sumamos 1
    if x[i]**2 + y[i]**2 <= 1:
        inside_count += 1
    # Calculamos la estimación de pi y la añadimos a la lista
    pi_estimate = 4 * inside_count / (i+1)
    pi_estimations.append(pi_estimate)

# ---- VISUALIZACIÓN DE PUNTOS EN EL CUADRADO/CÍRCULO ----
plt.figure(figsize=(8,8))
# Verifica qué puntos están dentro del círculo
inside_circle = x**2 + y**2 <= 1
# Y cuáles están fuera
outside_circle = x**2 + y**2 > 1
# Dibuja los puntos dentro del círculo en azul
plt.scatter(x[inside_circle], y[inside_circle], c='blue', marker='.', label='Dentro del Círculo')
# Y los puntos fuera del círculo en rojo
plt.scatter(x[outside_circle], y[outside_circle], c='red', marker='.', label='Fuera del Círculo')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Estimación de π usando Monte Carlo: π ≈ {pi_estimate:.4f}')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

# ---- VISUALIZACIÓN DE LA ESTIMACIÓN DE PI A LO LARGO DEL TIEMPO ----
plt.figure(figsize=(10,6))
# Dibuja cómo ha ido cambiando nuestra estimación de pi con más puntos
plt.plot(pi_estimations, '-o', markersize=2)
#'-o': Esta es una cadena que define el estilo del gráfico.
#El guion - indica que las estimaciones se conectarán con una línea, y la letra o indica que cada punto de datos (cada estimación de π en este caso)
# también se marcará con un círculo.
#markersize=2: Este es un argumento que define el tamaño del marcador (en este caso, los círculos o). El valor 2 significa que el tamaño de estos círculos será pequeño.
# Dibuja una línea mostrando el valor real de pi
plt.axhline(y=np.pi, color='r', linestyle='--', label='Valor real de π') #esta línea de código añade al gráfico una línea horizontal discontinua de color rojo en la posición y correspondiente al valor real de π y la etiqueta con el texto "Valor real de π
plt.title('Estimación de π con el tiempo')
plt.xlabel('Número de puntos')
plt.ylabel('Estimación de π')
plt.legend()
plt.grid(True)
plt.show()
```

Al ejecutar el código, verás una imagen con puntos azules (dentro del círculo) y puntos rojos (fuera del círculo). La estimación de π se muestra en el título. Cuantos más puntos uses (incrementando el valor de N), más precisa será la estimación, pero también llevará más tiempo calcularla.

El Juego de la Vida de Conway y sus Aplicaciones

Introducción

El Juego de la Vida es un autómata celular ideado por el matemático británico John Conway en 1970. Aunque es conocido como un "juego", en realidad no tiene jugadores en el sentido tradicional. Se juega en un tablero infinito bidimensional compuesto por "células", que pueden estar vivas o muertas. El juego evoluciona en pasos discretos, y el estado de cada célula en el siguiente paso se determina según reglas simples basadas en el número de células vecinas vivas.

1. Soledad: Si una célula viva tiene menos de 2 vecinas vivas, muere.
2. Estabilidad: Si una célula viva tiene 2 o 3 vecinas vivas, sigue viva.
3. Superpoblación: Si una célula viva tiene más de 3 vecinas vivas, muere.
4. Reproducción: Si una célula muerta tiene exactamente 3 vecinas vivas, se convierte en una célula viva.
Estas reglas determinan el estado de cada célula en el siguiente paso o generación basado en su estado actual y el de sus vecinos. La simplicidad de estas reglas es engañosa, ya que pueden dar lugar a patrones y estructuras extremadamente complejas y fascinantes.

numpy: Es una biblioteca para trabajar con matrices y realizar operaciones matemáticas de alto rendimiento.

```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors

# Definir la función que actualiza el estado del tablero del Juego de la Vida
def update(grid, N):
    # Crear una copia del tablero para no modificarlo mientras se calculan los cambios
    newGrid = grid.copy()

    # Recorrer cada celda del tablero
    for i in range(N):
        for j in range(N):
            # Sumar los valores de las celdas vecinas (tomando en cuenta los bordes del tablero con %N)
            total = (grid[i, (j-1)%N] + grid[i, (j+1)%N] +
                     grid[(i-1)%N, j] + grid[(i+1)%N, j] +
                     grid[(i-1)%N, (j-1)%N] + grid[(i-1)%N, (j+1)%N] +
                     grid[(i+1)%N, (j-1)%N] + grid[(i+1)%N, (j+1)%N])

            # Si la celda actual está viva
            if grid[i, j] == ON:
                # Aplicar reglas del Juego de la Vida para celdas vivas
                if (total < 2*ON) or (total > 3*ON):
                    newGrid[i, j] = OFF
            # Si la celda actual está muerta
            else:
                # Aplicar reglas del Juego de la Vida para celdas muertas
                if total == 3*ON :
                    newGrid[i, j] = ON

    # Actualizar el tablero original con los nuevos valores
    grid[:] = newGrid[:]

# Tamaño del tablero
N = 100
# Definir colores para las celdas: vivo, muerto y otro estado (rojo)
ON = 255    # Blanco
OFF = 0    # BLACK

vals = [ON, OFF]

# Generar una población inicial aleatoria para el tablero
grid = np.random.choice(vals, N*N, p=[0.1,  0.9]).reshape(N, N)

# Definir una paleta de colores personalizada para la visualización
cmap = matplotlib.colors.ListedColormap(['black', 'white'])

# Mostrar la evolución del Juego de la Vida paso a paso
for step in range(1):#!100
    # Configurar y mostrar el estado actual del tablero
    plt.figure(figsize=(6,6))
    plt.imshow(grid, cmap=cmap, interpolation='nearest')
    plt.title(f"Paso {step+1}")
    plt.axis('off')
    plt.show()

    # Actualizar el tablero para el siguiente paso
    update(grid, N)

```


# Simulación de la propagación de un incendio en un bosque.

El bosque se representa como una matriz donde cada celda puede tener uno de los siguientes estados:
- EMPTY (0): Representa un lugar vacío o tierra.
- TREE (1): Representa un árbol.
- BURNING (2): Representa un árbol que se está quemando.

Reglas de la simulación:
1. Si una celda está vacía y se cumple una probabilidad p, crecerá un árbol en esa celda.
2. Si una celda tiene un árbol y al menos uno de sus vecinos inmediatos está ardiendo, el árbol también comenzará a arder.
3. Si un árbol está ardiendo, al siguiente paso, se convertirá en tierra (estado vacío).
Se inicia la simulación con una densidad inicial de árboles y se prende fuego en el centro del bosque.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Definiciones de los estados de las celdas
EMPTY = 0
TREE = 1
BURNING = 2

def spread_of_fire(forest, p):
    """
    Avanza un paso en el tiempo siguiendo las reglas de propagación.
    """
    new_forest = forest.copy()
    for i in range(forest.shape[0]):
        for j in range(forest.shape[1]):
            if forest[i, j] == TREE:
                # Identificar vecinos
                neighbors = [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]

                # Si algún vecino está ardiendo, el árbol comienza a arder
                if any(forest[x, y] == BURNING for x, y in neighbors if 0 <= x < forest.shape[0] and 0 <= y < forest.shape[1]):
                    new_forest[i, j] = BURNING
                # Si se cumple la probabilidad p, crecerá un árbol
                elif np.random.rand() < p:
                    new_forest[i, j] = TREE
            elif forest[i, j] == BURNING:
                # Si el árbol está ardiendo, al siguiente paso se convierte en tierra
                new_forest[i, j] = EMPTY
    return new_forest

# Parámetros de la simulación
forest_size = (1, 1) #!100
initial_tree_density = 0.7
p = 0.01  # Probabilidad de que un árbol crezca en una celda vacía
steps = 1

# Iniciar el bosque con una densidad inicial de árboles
forest = np.where(np.random.rand(*forest_size) < initial_tree_density, TREE, EMPTY)

# Iniciar un fuego en el centro del bosque
forest[forest_size[0] // 2, forest_size[1] // 2] = BURNING

# Definir colores personalizados para la visualización
colors = ["black", "green", "red"]
cmap = ListedColormap(colors)

# Mostrar el estado inicial del bosque
plt.figure(figsize=(10, 10))
plt.imshow(forest, cmap=cmap, vmin=EMPTY, vmax=BURNING)
plt.colorbar(ticks=[EMPTY, TREE, BURNING])
plt.show()

# Avanzar en el tiempo y mostrar la evolución del bosque
for _ in range(steps):
    forest = spread_of_fire(forest, p)
    plt.figure(figsize=(10, 10))
    plt.imshow(forest, cmap=cmap, vmin=EMPTY, vmax=BURNING)
    plt.colorbar(ticks=[EMPTY, TREE, BURNING])
    plt.show()
```

TASEP (Totally Asymmetric Simple Exclusion Process) es un modelo matemático que se utiliza para describir sistemas con partículas (como autos en una carretera) que se mueven en una sola dirección, pero con una regla principal: dos partículas no pueden ocupar el mismo lugar al mismo tiempo. Es como una fila de autos en un carril: un auto no puede adelantar a otro, simplemente porque no hay espacio para hacerlo.

Modelo TASEP (Proceso de Exclusión Simple Totalmente Asimétrico):

¿Qué es?

Imagina una línea (o una carretera) con espacios donde pueden haber partículas (autos) o estar vacíos.

Las partículas solo se mueven en una dirección (por ejemplo, de izquierda a derecha) y siempre tratan de avanzar.

Reglas básicas:

Movimiento: Si el espacio siguiente está vacío, una partícula se mueve hacia ese espacio.
Exclusión: Si el espacio siguiente está ocupado por otra partícula, no puede moverse. ¡No hay adelantamientos!
Entrada y salida: En los extremos de la línea, las partículas pueden entrar con cierta probabilidad (si el primer sitio está vacío) y las partículas en el último sitio pueden salir, liberando ese sitio.
¿Para qué sirve?

Ayuda a entender cómo fluyen las cosas en sistemas donde no puede haber dos cosas en el mismo lugar al mismo tiempo.


Se utiliza en biología para estudiar cómo las moléculas se mueven dentro de las células.

También sirve para modelar sistemas de transporte y producción.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
L = 100  # Longitud del sistema (número de sitios)
rho = 0  # Densidad inicial de partículas
p = 1.0  # Probabilidad de que una partícula se mueva (si el siguiente sitio está vacío)
alpha = 0  # Probabilidad de entrada de una partícula en el sitio 1 (frontera izquierda)
beta = 0  # Probabilidad de salida de una partícula desde el último sitio (frontera derecha)

# Inicialización
sistema = np.where(np.random.rand(L) < rho, 1, 0)
evolucion = [sistema.copy()]

def actualiza_TASEP(s):
    """
    Realiza una actualización del sistema según las reglas del TASEP con fronteras abiertas.
    """
    # Frontera izquierda
    if s[0] == 0 and np.random.rand() < alpha:
        s[0] = 1



    # Movimientos internos
    for i in range(len(s) - 1, 0, -1):
        if s[i] == 0 and s[i-1] == 1 and np.random.rand() < p:
           s[i], s[i-1] = 1, 0

    # Frontera derecha
    if s[-1] == 1 and np.random.rand() < beta:
       s[-1] = 0

    return s

# Simulación
pasos = 200
for i in range(pasos):
    sistema = actualiza_TASEP(sistema)
    evolucion.append(sistema.copy())

# Visualización espacio-tiempo

plt.figure(figsize=(10, 10))
plt.imshow(evolucion, aspect='auto', cmap='binary')
plt.title(f"Evolución Espacio-Tiempo del TASEP con Fronteras Abiertas")
plt.ylabel("Tiempo")
plt.xlabel("Posición")
plt.colorbar(ticks=[0, 1], label='0: Vacío, 1: Partícula')
plt.show()
```

El modelo de Nagel-Schreckenberg (NaSch) es uno de los modelos más simples y conocidos para simular el tráfico vehicular en una carretera unidireccional. Fue propuesto en 1992 por Kai Nagel y Michael Schreckenberg y se ha utilizado ampliamente como un punto de partida para estudiar la dinámica del tráfico.

### **Descripción del Modelo Nagel-Schreckenberg**:

1. **Entorno**:
    - Consiste en una carretera unidireccional dividida en celdas, donde cada celda puede estar vacía o contener un vehículo.
    - Los vehículos se mueven de izquierda a derecha.
    - Usualmente, la carretera tiene condiciones de frontera periódicas, lo que significa que es circular: un vehículo que sale del extremo derecho entra nuevamente por el extremo izquierdo.

2. **Estado de los vehículos**:
    - Cada vehículo tiene una velocidad que es un número entero entre 0 y Vmax , donde Vmax  es la velocidad máxima permitida.

3. **Reglas de evolución (actualización en pasos discretos de tiempo)**:
    - **Aceleración**: Si la velocidad de un vehículo es menor que Vmax  y el número de celdas vacías delante de él es mayor que su velocidad, entonces aumenta su velocidad en una unidad.
    - **Desaceleración (frenado debido al tráfico)**: Si un vehículo tiene un número n de celdas vacías delante de él y n es menor que su velocidad, entonces reduce su velocidad a n.
    - **Aleatoriedad**: Con probabilidad p, reduce la velocidad del vehículo en una unidad (si su velocidad no es cero). Esto simula efectos aleatorios como el frenado debido a distracciones.
    - **Movimiento**: Cada vehículo se mueve hacia adelante tantas celdas como su velocidad.

### **Características y Fenomenología**:

- **Transiciones de Fase**:
    - A bajas densidades de vehículos, el tráfico fluye libremente. Pero a medida que la densidad aumenta, aparecen atascos espontáneos, y el flujo disminuye.
    - Hay una transición de una fase de flujo libre a una fase congestionada a medida que la densidad de vehículos cruza un umbral crítico.

- **Formación de Atascos**:
    - Los atascos pueden surgir espontáneamente incluso en ausencia de obstáculos. Estos atascos se mueven en dirección opuesta al tráfico y son una característica distintiva del modelo.

### **Aplicaciones**:

1. **Estudio de Transiciones de Fase**: El modelo NaSch ha sido instrumental en mostrar que el tráfico puede exhibir transiciones de fase, similares a las transiciones entre estados sólido, líquido y gaseoso en la física.
2. **Bases para Modelos más Complejos**: Es una base sobre la cual se han propuesto variantes más sofisticadas que consideran múltiples carriles, interacciones más complejas y otros aspectos del tráfico.
3. **Optimización del Tráfico**: A través de la comprensión de los patrones de tráfico, es posible proponer estrategias para optimizar el flujo y reducir los atascos.

En resumen, el modelo de Nagel-Schreckenberg ha sido fundamental en la física del tráfico y ha proporcionado una visión profunda de cómo los comportamientos locales (como el frenado de un automóvil) pueden llevar a fenómenos colectivos emergentes (como la formación de atascos de tráfico).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del modelo
L = 100  # Longitud de la carretera (número de celdas)
T = 100  # Número total de pasos de tiempo
density = 0.3  # Densidad de coches en la carretera
vmax = 5  # Velocidad máxima en términos de celdas por paso de tiempo
p = 0.1  # Probabilidad de frenado aleatorio

# Inicializar carretera con -1 (celda vacía) y vehículos (0 <= v <= vmax)
road = -np.ones(L, dtype=int)
initial_positions = np.random.choice(range(L), int(density*L), replace=False)
road[initial_positions] = np.random.randint(0, vmax+1, size=len(initial_positions))

# Para almacenar el historial de la carretera a lo largo del tiempo
history = [road.copy()]

for t in range(T):
    # Aplicar reglas
    next_road = -np.ones(L, dtype=int)
    for i in range(L):
        if road[i] != -1:
            v = road[i]
            # Aceleración
            v = min(v + 1, vmax)
            # Desaceleración
            d = 1
            while d <= v and road[(i + d) % L] == -1:
                d += 1
            v = d - 1
            # Randomización
            if np.random.rand() < p:
                v = max(v - 1, 0)

            # Mover coche
            next_pos = (i + v) % L
            next_road[next_pos] = v

    road = next_road.copy()
    history.append(road)

# Visualización en 2D
plt.figure(figsize=(10,10))
plt.imshow(np.array(history) != -1, cmap='gray_r', origin='lower', aspect='auto')
plt.colorbar(label="Presencia de coche")
plt.xlabel("Posición en la carretera")
plt.ylabel("Tiempo")
plt.title("Diagrama Espacio-Tiempo del Modelo de Nagel-Schreckenberg")
plt.show()
```


Este código simula el modelo Nagel-Schreckenberg, un modelo sencillo para tráfico vehicular
en una carretera unidimensional. El modelo describe el movimiento de los vehículos considerando
las velocidades y la distancia entre vehículos.

Los parámetros principales son:
- L: Longitud total de la carretera (o número total de sitios).
- tiempomax: Tiempo máximo de simulación.
- tiempocalc: Tiempo a partir del cual se empieza a calcular la densidad y la velocidad.
- MOY: Número de veces que se realiza la simulación.
- PN: Probabilidad de que un vehículo disminuya su velocidad.
- IVMAX: Velocidad máxima permitida.

Los resultados muestran cómo cambia la velocidad y el flujo de vehículos en función de la densidad de la carretera.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
L = 100
tiempomax= 3000
tiempocalc = 2000
MOY = 1
PN = 0
IVMAX = 3


# For random number generation
np.random.seed(10)

def init(IR):
    pos = []
    for i in range(len(IR)):
        if IR[i] == 1:
            pos.append(i)
    return pos

def movimiento(IR, IV, PN, IVMAX):
    IPR = sum(IR)
    ICR = 0

    pos = init(IR)

    for i in range(IPR):
        k = pos[i]

        if i != IPR - 1:
            gap = pos[i+1] - k - 1
        else:
            gap = L + pos[0] - k - 1

        v_p = IV[k] + 1
        IV[k] = min(v_p, IVMAX, gap)
        v_m = IV[k] - 1

        if PN > np.random.rand():
            IV[k] = max(v_m, 0)

        ICR += IV[k]

        if IV[k] > 0:
            IR[k] = 0
            kf = (k + IV[k]) % L
            IR[kf] = 1
            IV[kf] = IV[k]
            IV[k] = 0

    return ICR


def main():
    densities = []
    flow_rates = []
    velocities = []


    for iiden in range(101):
        den = iiden / 100.0
        cr = 0
        velo = 0

        for imoy in range(MOY):
            IR = np.zeros(L, dtype=int)
            IV = np.zeros(L, dtype=int)
            iden = int(round(L * den))
            positions = np.random.choice(range(L), iden, replace=False)
            IR[positions] = 1

            for tiempo in range(tiempomax):
                icr = movimiento(IR, IV, PN, IVMAX)
                if tiempo > tiempocalc:
                    if sum(IR) != 0:
                        corr = icr / float(sum(IR))
                        cor = icr / float(L)
                    else:
                        corr = 0
                        cor = 0
                    cr += cor
                    velo += corr

        cr /= (MOY * (tiempomax - tiempocalc))
        velo /= (MOY * (tiempomax - tiempocalc))

        densities.append(den)
        flow_rates.append(cr)
        velocities.append(velo)
    #print(den,cr,velo)

    plt.figure()
    plt.plot(densities, flow_rates, label='Flow Rate')
    plt.plot(densities, velocities, label='Velocity')
    plt.legend()
    plt.xlabel('Density')
    plt.ylabel('Value')
    plt.title('Nagel-Schreckenberg Model')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```

El modelo BML (Biham-Middleton-Levine) es un modelo de tráfico de partículas en una red cuadriculada, diseñado para estudiar el flujo de tráfico en condiciones de congestión. El modelo toma su nombre de Ofer Biham, A. Alan Middleton y Dov Levine, quienes lo introdujeron en 1992.

### **Descripción del Modelo BML**:

1. **Entorno**:
    - Una red cuadriculada (o malla) de tamaño \(L \times L\). Cada celda de la malla puede estar vacía o contener un vehículo.
  
2. **Vehículos**:
    - Hay dos tipos de vehículos: aquellos que se mueven solo en dirección horizontal (denominados a menudo "vehículos tipo 1" o "vehículos rojos") y aquellos que se mueven solo en dirección vertical (denominados "vehículos tipo 2" o "vehículos azules").

3. **Reglas de movimiento**:
    - En cada paso de tiempo, todos los vehículos intentan moverse una celda en su dirección preferida (los vehículos rojos intentan moverse a la derecha y los vehículos azules intentan moverse hacia arriba).
    - Un vehículo solo puede moverse si la celda en su dirección preferida está vacía.
    - Los vehículos de un tipo dado (por ejemplo, los rojos) intentan moverse simultáneamente en un paso de tiempo. Luego, en el siguiente paso de tiempo, los vehículos del otro tipo (en este caso, los azules) intentan moverse. Los dos tipos de vehículos se alternan en sus intentos de moverse.

4. **Condiciones de frontera**:
    - Las condiciones de frontera son periódicas. Esto significa que un vehículo que se mueve fuera del borde derecho de la malla reaparecerá en el borde izquierdo (y viceversa). Lo mismo ocurre en la dirección vertical: un vehículo que se mueve fuera del borde superior reaparecerá en el borde inferior.

### **Características y Fenomenología**:

- **Transición de Fase**:
    - Una de las características más interesantes del modelo BML es su transición de fase. A bajas densidades de vehículos, el tráfico fluye libremente sin obstrucciones. Sin embargo, a medida que la densidad de vehículos aumenta, se llega a un punto crítico en el que el tráfico se congestiona y se forma un atasco que abarca todo el sistema.

- **Formación de Patrones**:
    - En ciertas densidades, el modelo BML muestra la formación de patrones y estructuras interesantes. Por ejemplo, pueden surgir "avenidas" libres de tráfico en ciertas direcciones, mientras que otras direcciones están completamente bloqueadas.

### **Aplicaciones**:

Aunque el modelo BML es una simplificación abstracta del tráfico real, ha sido útil para:
1. Estudiar las transiciones de fase y la formación de patrones en sistemas fuera del equilibrio.
2. Obtener intuiciones sobre cómo las interacciones locales pueden llevar a comportamientos colectivos emergentes.
3. Investigar cómo las perturbaciones o cambios en las reglas pueden influir en la dinámica del tráfico.

En resumen, el modelo BML ofrece un marco simple pero poderoso para estudiar la dinámica colectiva en sistemas con muchas partes interactuantes, y ha inspirado investigaciones en diversas áreas, desde la física estadística hasta la investigación de operaciones y la ingeniería de tráfico.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Parámetros
L = 50  # Tamaño de la malla
density = 0.8  # Densidad de coches

# Colores: blanco para vacío, rojo para coches tipo A (left-right) y azul para coches tipo B (up-down)
colors = ['white', 'red', 'blue']
cmap = ListedColormap(colors)

# Inicialización
grid = np.zeros((L, L), dtype=int)  # 0: vacío, 1: coche tipo A, 2: coche tipo B
initial_positions_A = np.random.choice(range(L*L), int(density/2 * L*L), replace=False)
initial_positions_B = np.random.choice(list(set(range(L*L)) - set(initial_positions_A)), int(density/2 * L*L), replace=False)
for pos in initial_positions_A:
    grid[pos//L][pos%L] = 1
for pos in initial_positions_B:
    grid[pos//L][pos%L] = 2

def update(grid):
    # Movimiento de coches tipo A
    next_grid = grid.copy()
    for i in range(L):
        for j in range(L):
            if grid[i][j] == 1:
                if grid[i][(j+1)%L] == 0:  # Verificar si la celda a la derecha está vacía
                    next_grid[i][(j+1)%L] = 1
                    next_grid[i][j] = 0

    # Movimiento de coches tipo B
    grid = next_grid.copy()
    for i in range(L):
        for j in range(L):
            if next_grid[i][j] == 2:
                if next_grid[(i+1)%L][j] == 0:  # Verificar si la celda de arriba está vacía
                    grid[(i+1)%L][j] = 2
                    grid[i][j] = 0

    return grid

# Simulación y visualización
plt.figure(figsize=(10, 10))
for step in range(100):
    grid = update(grid)
    plt.imshow(grid, cmap=cmap, origin='upper')
    plt.title(f"Paso {step+1}")
    plt.pause(0.1)
    plt.clf()

plt.show()

```
